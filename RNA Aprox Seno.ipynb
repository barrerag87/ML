Esta red neuronal tiene una capa oculta con 10 neuronas y utiliza la funcion de activacion ReLU. La red neuronal se entrena con 100 epocas y se evalua con 1000 puntos. La grafica muestra que la red neuronal aproxima la funcion seno con una buena precision.

```python
import numpy as np
import tensorflow as tf

# Definimos la funcion seno
def seno(x):
  return np.sin(x)

# Creamos los datos de entrenamiento
x_train = np.linspace(-np.pi, np.pi, 100)
y_train = seno(x_train)

# Definimos la red neuronal
model = tf.keras.Sequential([
  tf.keras.layers.Dense(10, activation='relu', input_shape=(1,)),
  tf.keras.layers.Dense(1)
])

# Compilamos la red neuronal
model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])

# Entrenamos la red neuronal
model.fit(x_train, y_train, epochs=100)

# Evaluamos la red neuronal
x_test = np.linspace(-np.pi, np.pi, 1000)
y_test = seno(x_test)
y_pred = model.predict(x_test)

# Imprimimos la grafica
import matplotlib.pyplot as plt

plt.plot(x_test, y_test, label='Real')
plt.plot(x_test, y_pred, label='Prediccion')
plt.legend()
plt.show()
```

```python
import numpy as np
import tensorflow as tf

# Definimos la funcion
def f(x):
  return 2*x**2 - x**3

# Creamos los datos de entrenamiento
x_train = np.linspace(-1, 1, 100)
y_train = f(x_train)

# Definimos la red neuronal
model = tf.keras.Sequential([
  tf.keras.layers.Dense(10, activation='relu', input_shape=(1,)),
  tf.keras.layers.Dense(1)
])

# Compilamos la red neuronal
model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])

# Entrenamos la red neuronal
model.fit(x_train, y_train, epochs=100)

# Evaluamos la red neuronal
x_test = np.linspace(-1, 1, 1000)
y_test = f(x_test)
y_pred = model.predict(x_test)

# Imprimimos la grafica
import matplotlib.pyplot as plt

plt.plot(x_test, y_test, label='Real')
plt.plot(x_test, y_pred, label='Prediccion')
plt.legend()
plt.show()
```

```python
import numpy as np
import tensorflow as tf

# Define the function
def f(x):
  return 2 * x**2 - x**3

# Create training data
x_train = np.linspace(-1.5, 1.5, 2000)  # Increased data points
y_train = f(x_train)

# Define the neural network
model = tf.keras.Sequential([
  tf.keras.layers.Dense(64, activation='relu', input_shape=(1,)),  # More neurons
  tf.keras.layers.Dropout(0.2),  # Introduce dropout for regularization
  tf.keras.layers.Dense(128, activation='relu'),  # More neurons
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(1)
])

# Compile the neural network with advanced optimizer
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='mse', metrics=['accuracy'])

# Train the neural network with more epochs
model.fit(x_train, y_train, epochs=500)

# Evaluate the neural network with denser test data
x_test = np.linspace(-1.5, 1.5, 10000)
y_test = f(x_test)
y_pred = model.predict(x_test)

# Plot the results
import matplotlib.pyplot as plt

plt.plot(x_test, y_test, label='Real')
plt.plot(x_test, y_pred, label='Prediction')
plt.legend()
plt.show()
```
